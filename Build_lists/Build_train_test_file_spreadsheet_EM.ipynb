Training, Validation and testing dataset were labeled by batch manually

```python
import sys
sys.path.append('/host/d/Github/')
import os
import numpy as np
import pandas as pd
import nibabel as nb
import Diffusion_denoising_thin_slice.functions_collection as ff
```

```python
patient_sheet = ff.find_all_target_files(['*'],os.path.join('/host/d/Data/minnie_EM/nii'))
print('total patient num: ', len(patient_sheet))
```

```python
# 
results = []
for i in range(0, len(patient_sheet)):
    patient_id = os.path.basename(patient_sheet[i])
    # first 6 are train, seven and eight are val, last three are test
    if i < 17:
        batch = 'train'
    elif i>=17 and i<20:
        batch = 'val'
    elif i>=20:
        batch = 'test'
    results.append([batch, patient_id])
df = pd.DataFrame(results, columns=['batch','Patient_ID'])
save_path = os.path.join('/host/d/Data/minnie_EM/Patient_lists','minnie_EM_batched.xlsx')
df.to_excel(save_path, index=False)
```

# build list for simulations

```python
patient_sheet = pd.read_excel(os.path.join('/host/d/Data/minnie_EM/Patient_lists/minnie_EM_batched.xlsx'),dtype={'Patient_ID': str})
noise_types = ['gaussian']

data_path = '/host/d/Data/minnie_EM'

for noise_type in noise_types:
    results = []
    for i in range(0, len(patient_sheet)):
        patient_id = patient_sheet['Patient_ID'][i]
        batch = patient_sheet['batch'][i]
        print(f"Processing patient {patient_id} in batch {batch} with noise type {noise_type}")


        ground_truth_file = os.path.join(data_path,'nii', patient_id, 'img.nii.gz')
        img = nb.load(ground_truth_file).get_fdata()
        slice_num = img.shape[2]
        x_shape, y_shape = img.shape[0], img.shape[1]
        max_value = np.max(img)
        min_value = np.min(img)
        print('ground truth max and min value:', max_value, min_value)

        n_type = 'gaussian' if noise_type == 'gaussian' else 'poisson'
        simulation_file_1 = os.path.join(data_path,'simulation_v1', patient_id, 'recon_1', 'img.nii.gz')
        simulation_file_2 = os.path.join(data_path,'simulation_v1', patient_id, 'recon_2', 'img.nii.gz')
        
        results.append([batch,patient_id, simulation_file_1, simulation_file_2, ground_truth_file, slice_num, x_shape, y_shape])

        df = pd.DataFrame(results, columns=['batch','Patient_ID', 'simulation_file_1','simulation_file_2', 'ground_truth_file', 'slice_num', 'x_shape', 'y_shape'])
        df.to_excel(os.path.join('/host/d/Data/minnie_EM/Patient_lists', 'minnie_EM_split_gaussian_simulation_v1.xlsx'), index=False)
```
