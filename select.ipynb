```python
import sys
sys.path.append('/host/d/Github/')
import nibabel as nb
import glob
import os
import numpy as np
import shutil
import Diffusion_denoising_thin_slice.functions_collection as ff
```

```python
trial = 'unsupervised_gaussian_2'
pick_epoch = 190
main_path = '/host/d/projects/denoising/models'
data_path = os.path.join(main_path, trial, 'pred_images_input_both')
```

### rename

```python
patient_list = ff.find_all_target_files(['*/random_*'], data_path)
for i in range(len(patient_list)):
    # # not our method:
    # epoch_folder = os.path.join(patient_list[i], 'epoch' + str(pick_epoch)+'_1')
    # new_folder_name = os.path.join(patient_list[i], 'sample')
    # # copy epoch folder to sample folder
    # if os.path.exists(epoch_folder):
    #     shutil.rmtree(new_folder_name, ignore_errors=True)
    # if not os.path.exists(new_folder_name):
    #    shutil.copytree(epoch_folder, new_folder_name)

    # our method:
    sample_name = 'sample2'
    epoch_folders = ff.sort_timeframe(ff.find_all_target_files(['epoch'+str(pick_epoch)+'_*'], patient_list[i]),0,'_')
    for j in range(len(epoch_folders)):
        times = ff.find_timeframe(epoch_folders[j],0,'_')
        new_folder_name = os.path.join(patient_list[i], sample_name+'_'+str(times))
        ff.make_folder([new_folder_name])
        image_files = ff.find_all_target_files(['*.nii.gz'], epoch_folders[j])
        for k in range(len(image_files)):
            if not os.path.exists(os.path.join(new_folder_name, os.path.basename(image_files[k]))):
                shutil.copy(image_files[k], os.path.join(new_folder_name, os.path.basename(image_files[k])))

    avg_folder = os.path.join(patient_list[i], 'epoch' + str(pick_epoch)+'avg')
    new_avg_folder_name = os.path.join(patient_list[i], sample_name+'avg')
    ff.make_folder([new_avg_folder_name])
    image_files = ff.find_all_target_files(['*.nii.gz'], avg_folder)
    for k in range(len(image_files)):
        if not os.path.exists(os.path.join(new_avg_folder_name, os.path.basename(image_files[k]))):
            shutil.copy(image_files[k], os.path.join(new_avg_folder_name, os.path.basename(image_files[k])))
    
```

### delete

```python
patient_list = ff.find_all_target_files(['*/random_*'], data_path)
for i in range(len(patient_list)):
    
    epoch_folders = ff.find_all_target_files(['sam*'], patient_list[i])
    for j in range(len(epoch_folders)):
        shutil.rmtree(epoch_folders[j], ignore_errors=True)
```
