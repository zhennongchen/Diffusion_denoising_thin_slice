### task: exclude the blank slices

```python
from turtle import left


patient_list = pd.read_csv('/host/d/Data/PCCT/Patient_lists/raw_info.csv',dtype={'Patient_ID':str})
patient_list = patient_list[patient_list['NumChannels'] == 1]

for i in range(0,len(patient_list)):
    patient_id = patient_list['Patient_ID'].iloc[i]
    data_path = os.path.join('/host/d/Data/PCCT/data/soft_thins', patient_id, 'soft_thins_0.nii.gz')

    # if os.path.exists(os.path.join('/host/d/Data/PCCT/data/soft_thins', patient_id, 'soft_thins_0_noblank.nii.gz')):
    #     print(f"Patient {patient_id} already processed, skip.")
    #     continue
    print(f"Processing patient {patient_id}...")
    img = nb.load(data_path)
    affine = img.affine
    data = img.get_fdata()
    print('data shape: ', data.shape)
    
    # nonblack slices means the number of pixels with value >0 has more than 20% of total pixels
    # non_blank_slices = [i for i in range(data.shape[2]) if np.sum(data[:,:,i]>-500) > (0.2*data.shape[0]*data.shape[1])]
    # print('non blank slices: ', non_blank_slices)
    # min_slice = min(non_blank_slices)
    # max_slice = max(non_blank_slices)
    # non_blank_slices = list(range(min_slice, max_slice+1))
    # print('orignally we have ', data.shape[2], 'slices, after excluding blank slices we have ', len(non_blank_slices), 'slices.')

    # find the middle slice and keep 50 left and 50 right slices
    slice_num = data.shape[2]
    mid_slice = slice_num // 2
    left_bound = max(0, mid_slice - 50)
    right_bound = left_bound + 100
    assert right_bound <= slice_num, "Right bound exceeds number of slices."
    non_blank_slices = list(range(left_bound, right_bound))

    new_data = data[:,:,non_blank_slices]

    nb.save(nb.Nifti1Image(new_data, affine), os.path.join('/host/d/Data/PCCT/data/soft_thins', patient_id, 'soft_thins_0_noblank.nii.gz'))
```
