### diffusion illustration

```python
data_path = '/host/d/Data/brain_CT/simulation'
patient_id = '00214841'
patient_subid = '0000455418'
save_folder = '/host/d/projects/denoising/pictures/noisy_noisy_pairs/brain_CT_'+patient_id
ff.make_folder([save_folder])

slice_i = 54
ii = nb.load(os.path.join(data_path, patient_id, patient_subid, 'gaussian_random_0/recon.nii.gz')).get_fdata()[:,:,slice_i]

ii = Data_processing.cutoff_intensity(ii,cutoff_low = -1000, cutoff_high = 2000)
ii = Data_processing.normalize_image(ii, normalize_factor = 'equation', image_max = 2000, image_min = -1000 ,invert = False)

gaussian_noise = np.random.normal(0, 0.1, ii.shape)

noisy_ii = ii + 0.25*gaussian_noise
noisy_ii = Data_processing.normalize_image(noisy_ii, normalize_factor = 'equation', image_max = 2000, image_min = -1000 ,invert = True)

ff.save_grayscale_image(np.flip(noisy_ii.T,0),os.path.join(save_folder, 'noisy_slice' + str(slice_i) + '_altered.png'), normalize = True, WL = 40, WW = 80)
```
