```python
import os
import numpy as np
import glob
import shutil
```

```python
def find_all_target_files(target_file_name,main_folder):
    F = np.array([])
    for i in target_file_name:
        f = np.array(sorted(glob.glob(os.path.join(main_folder, os.path.normpath(i)))))
        F = np.concatenate((F,f))
    return F
```

```python
local_path = '/Volumes/WD/Data_for_WMA/original-nii-images/'
folders = find_all_target_files(['abnormal_original_nii_2020/*'], local_path)
print('folders length',     len(folders))
```

```python
nas_path = '/Volumes/Camca/home/ZC/4DCT/nii-images/'
```

```python
for i in range(len(folders)):
    folder = folders[i]
    patient_class = 'Abnormal'#os.path.basename(os.path.dirname(folder))
    patient_id = os.path.basename(folder)

    des_folder = os.path.join(nas_path, patient_class, patient_id)
    print(patient_class, patient_id)
    os.makedirs(des_folder, exist_ok=True)
    os.makedirs(os.path.join(des_folder, 'img-nii'), exist_ok=True)
    
    files = find_all_target_files(['img-nii/*.nii.gz'], folder)

    for file in files:
        if os.path.isfile(os.path.join(des_folder, 'img-nii', os.path.basename(file))) == False:
            shutil.copy(file, os.path.join(des_folder, 'img-nii', os.path.basename(file)))
```
